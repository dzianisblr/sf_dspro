```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df['Date'] = pd.to_datetime(melb_df['Date'])

execute_list = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
for col in melb_df.columns:
    if melb_df[col].nunique() < 150 and col not in execute_list:
        melb_df[col] = melb_df[col].astype('category')
        
melb_df.head()
```

```python
melb_df.info()
```

```python
melb_df.groupby('Rooms')['Price'].mean()
```

```python
melb_df.groupby('Regionname')['Lattitude'].std().sort_values()
```

```python
selling_2017 = melb_df[(melb_df['Date']>= '2017-05-01') & (melb_df['Date']<= '2017-09-01')]
selling_2017.groupby('SellerG')['Price'].sum().sort_values().head(1)
```

```python
melb_df.pivot_table(
    values='BuildingArea',
    index='Type',
    columns='Rooms',
    aggfunc='median'
)
```

```python
mean_price_by_type = melb_df.pivot_table(
    values='Price',
    index='SellerG',
    columns='Type',
)
mean_price_by_type.sort_values('unit')
```
