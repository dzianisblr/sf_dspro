```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
```

```python
date_quarter = melb_df['Date'].dt.quarter
date_quarter.value_counts()
```

```python
execute_list = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
for col in melb_df.columns:
    if melb_df[col].nunique() < 150 and col not in execute_list:
        melb_df[col] = melb_df[col].astype('category')
melb_df.info()
```

```python
melb_df.sort_values(by='Price').head()
```

```python
melb_df.sort_values(by='Date', ascending=False)
```

```python
melb_df.sort_values(by=['Price', 'Distance']).loc[::10, ['Distance', 'Price']]
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df.sort_values(
    by=['AreaRatio'], 
    ignore_index=True, 
    ascending=False
    ).loc[1558, 'BuildingArea']
```

```python
town_tabl = melb_df[(melb_df['Rooms']>2) & (melb_df['Type']== 'townhouse')]
town_tabl.sort_values(
    by=['Rooms', 'MeanRoomsSquare'], 
    ascending=[True, False], 
    ignore_index=True
).loc[18, 'Price']
```

```python
melb_df.groupby(by='Type').mean()
```

```python
melb_df.groupby('Type')['Price'].mean()
```

```python
melb_df.groupby('Regionname')['Distance'].min().sort_values(ascending=False)
```

```python
melb_df.groupby('MonthSale')['Price'].agg(
    ['count', 'mean', 'max']
).sort_values(by='count', ascending=False)
```

```python
melb_df.groupby('Regionname')['SellerG'].agg(
    		['nunique', set]
)
```

```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
pivot
```

```python
movies = pd.read_csv('data/movies.csv')
ratings1 = pd.read_csv('data/ratings1.csv')
ratings2 = pd.read_csv('data/ratings2.csv')
dates = pd.read_csv('data/dates.csv')
```

```python
ratings = pd.concat(
    [ratings1, ratings2],
    ignore_index=True
)
ratings
```

```python
print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', dates.shape[0])
print(ratings.shape[0] == dates.shape[0])
```

```python
ratings_dates = pd.concat([ratings, dates], axis=1)
ratings_dates.tail(7)
```

```python
joined_false = ratings_dates.join(
    movies,
    rsuffix='_right',
    how='left'
)
joined_false
```

```python
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
joined.head()
```

```python
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
merged.head()
```

```python
merged2 = ratings_dates.merge(
    movies,
    on='movieId',
    how='outer'
)
print('Число строк в таблице merged2: ', merged2.shape[0])
merged2.tail()
```
